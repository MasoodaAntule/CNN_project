```python

# importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# advance libraries

import keras
from keras.datasets import mnist
```

```python

(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
# check the size of data
print(x_train.shape)
print(y_train.shape)
print("* "*10)
print(x_test.shape)
print(y_test.shape)
```

# Changing the dimension

```python
x_train.ndim
```

```python
x_train.shape
```

```python
x_train = np.expand_dims(x_train,-1)

x_test = np.expand_dims(x_test,-1)
```

```python
print(x_train.shape)

print(x_test.shape)
```

```python
x_train[0]
```

```python
x_train = x_train/255

x_test = x_test/255
```

```python
x_train[0]
```

# Splitting The Dataset

```python
x_train, x_val, y_train, y_val = train_test_split(
                                            x_train,
                                            y_train,
                                            random_state=42,
                                            test_size=0.2
)
```

```python
x_train.shape
```

```python
x_val.shape
```

# Building the Convolution Neural Network

```python
model = keras.models.Sequential([
                    keras.layers.Conv2D(filters=32,kernel_size=3,strides=(1,1),padding="valid",activation="relu",input_shape=[28, 28, 1]),
                    keras.layers.MaxPooling2D(pool_size=(2,2)),
                    keras.layers.Conv2D(filters=12,kernel_size=3,strides=(1,1),padding="valid",activation="relu"),
                    keras.layers.MaxPooling2D(pool_size=(2,2)),
                    keras.layers.Flatten(),
                    keras.layers.Dense(units=128,activation="relu"),
                    keras.layers.Dense(units=50,activation="relu"),
                    keras.layers.Dense(units=10,activation="softmax")
])
```

```python
model.summary()
```

```python
model.compile(
          optimizer="adam",
          loss="sparse_categorical_crossentropy",
          metrics=["accuracy"]
)
```

```python
model.fit(
    x_train,
    y_train,
    epochs=30,
    batch_size=512,
    verbose=True,
    validation_data=(x_val,y_val)
)

```

# Testing the model accuracy on the testing data


```python
np.argmax(model.predict(np.expand_dims(x_test[0],axis=0))).round(2)
```

```python
y_pred = model.predict(x_test)

y_pred
```

```python
model.evaluate(x_test,y_test)
```

#Plotting the confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python

plt.figure(figsize=(15,10))

y_pred_lables = [np.argmax(label) for label in y_pred]

cnf = confusion_matrix(y_test,y_pred_lables)

sns.heatmap(cnf,annot=True,fmt="d",xticklabels=class_labels,yticklabels=class_labels)
```

```python

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python

accuracy = accuracy_score(y_test, y_pred_lables)
print(f'Accuracy: {accuracy}')
```
